2.2 Extraed los valores únicos su frecuencia. BONUS presenta los resultados en un dataframe. 

Donde variable es cada una de las variables categóricas que tenemos en el dataset y n_unicos es una lista con los nombres de los valores únicos de cada variable.


```python
df_categoricas["case_number"].unique() 
```

```python
variables = df_categoricas.columns #Sacamos los nombres de las columnas categóricas y las almacenamos en la variable variables
```

```python
n_uniques= [] #Creamos una lista vacía para almacenar los n_uniques
for col in variables: #Iteramos por las columnas de las columnas de df_categoricas y por cada una apendeamosa la lista n_uniques el valor del unico de cada columna 
    n_uniques.append(list(df_categoricas[col].unique())) # lista de los n_uniques 
```

```python
dicc = {'variable': variables, 'n_uniques': n_uniques} #Convertimos las dos listas que tenemos en un diccionario en el que la key es variable y n_uniques y los values son los nombres de las columnas y su lista de valores unicos
```

```python
dicc
```

```python
final = pd.DataFrame.from_dict(dicc) #Convertimos el diccionario en un dataframe
```

```python
final.head() #tenemos el dataframe con las variables y los valores únicos
```
